```python
import numpy as np
```

```python
def squares (values):
    result = [ ]
    for v in values:
        result.append(v * v)
    return result
```

```python
to_square = range(100000)
%timeit squares(to_square)
```

```python
array_to_square = np.arange(0, 100000)
```

```python
%timeit array_to_square **2
```
